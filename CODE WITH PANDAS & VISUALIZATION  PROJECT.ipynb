```python
# FROM HERE I'M TAKING A NEW CSV FILE FOR FURTHER ANALYSIS..... CSV FILE NAME IS ("drinks.csv")

df=pd.read_csv("drinks.csv") # itss all about drinks
df
```

```python
df.info() # there can be null value as (?)
```

```python
df.describe()  # skewed in beer__serving  nd little skewedness in total_litres_of_pure_alcohol
```

```python
#univariate analysis / individual analysis column
#
```

```python
#univariate analysis
df['continent'].hist(figsize=(10,5),color='red')    

# In data there is 50+ continent in africa
#the lowest is a south america 10 to 15
#there was a small difference between asia continent and Europe continent
```

```python
df[['beer_servings','spirit_servings','wine_servings']].hist(figsize=(10,10),color='purple')
plt.tight_layout()
plt.show()  
#in 0 to 50 there was higher no of beer servings and small no of lower beer servings also there.
#in 300 to 350 there was lower no of beer servings in that 300 to 350 there was small no of higher beer servings also there.

#in spirit servings 0 to 100 there was higher no of spirit servings and there small no of lower spirit servings also there.
#300 to 400 lower no of spirit servings. In this there to much small higher no of spirit servings also there.

#in wine servings there was high no wine servings in 0 to 50 is 120+
#300 to 350 lower no of wine servings. 
```

```python
mtotal_litres_of_pure_alcohol=df['total_litres_of_pure_alcohol'].mean()
metotal_litres_of_pure_alcohol=df['total_litres_of_pure_alcohol'].median()

sns.distplot(df['total_litres_of_pure_alcohol'])
plt.axvline(mtotal_litres_of_pure_alcohol,color='g')
plt.axvline(metotal_litres_of_pure_alcohol,color='y')
plt.grid(True)

plt.show()

#left dn side is positive side and high chances of skewed is there & right nd side is lower
```

```python
sns.countplot(df['continent'])
plt.show() 

#there is a higher total_litres_of_pure_alcohol in Africa continent
#there is a lower total_litres_of_pure_alcohol in south America
```

```python
df['continent'].value_counts().plot(figsize=(7,7),kind='pie',autopct='%.2f%%')

#According to the given pie chart , we can clearly analize the proportion of selling alcohol-
#percentage with respect to its givenÂ countryÂ names.

#In africa there was higher no alcohol is sale 27.46%
#In South america there was lower no alcohol is sale 6.22%
#In europe and asia continent there was small no differences in selling of alcohol as compared to-
#other continents.
```

```python
#bivariate analysis
```

```python
#scatter plot 
plt.figure(figsize=(8,8))
sns.scatterplot(data=df,x='continent',y='total_litres_of_pure_alcohol')    
plt.show()
# In europe there was high no of total litres of pure alcohol
#In south america there was low no of total litres of pure alcohol
```
